```python
import os
import numpy as np
import pandas as pd
from glob import glob
import geopandas as gpd
from system_process import makedirs, copy_file
from raster_process import read_raster_arr_object, write_array_to_raster, clip_resample_reproject_raster
```

```python
WestUS_county = '../Data/shapefiles/Western_US/WestUS_county.shp'
county_df = gpd.read_file(WestUS_county)
county_df['fips'] = county_df['STATEFP'] + county_df['COUNTYFP']
county_df['fips'] = county_df['fips'].astype('int64')
print(county_df.info())
county_df.plot()
```

```python
wateruse_data = glob(os.path.join('../Data/USGS_water_use_data', '*2010.xlsx'))

for data in wateruse_data:
    wateruse_df = pd.read_excel(data, sheet_name='CountyData')
    wateruse_df = wateruse_df[['COUNTY', 'COUNTYFIPS', 'FIPS', 'YEAR', 'TO-WGWFr']]

wateruse_df
```

```python
county_wateruse = county_df.merge(wateruse_df, left_on='fips', right_on = 'FIPS', how='inner')
county_wateruse
```

```python
county_wateruse['gw_withdrawal'] = 1e6*1000*county_wateruse['TO-WGWFr']*0.00378541*365/county_wateruse['ALAND']  # convert to mm/year
county_wateruse['gw_withdrawal']
```

```python
county_wateruse.plot(column='gw_withdrawal', vmin=0, vmax=300)
```

```python
state = '../Data/shapefiles/Western_US/WestUS_states.shp'
state_df =gpd.read_file(state)


base = county_wateruse.plot(column='gw_withdrawal', vmin=0, vmax=300)
state_df.plot(ax=base, facecolor='none', edgecolor='red')
```

```python
base = county_wateruse[county_wateruse['gw_withdrawal']>50].plot(column='gw_withdrawal', vmin=0, vmax=300)
state_df.plot(ax=base, facecolor='none', edgecolor='red')
```
