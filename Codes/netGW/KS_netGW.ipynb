```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv('../../scratch/netGW_KS/KS_GMD4_netGW_pumping.csv')
df
```

```python
r2 = r2_score(df.KS_pump_mm, df.KS_netGW_mm)
print('R2: ', r2)

rmse = mean_squared_error(y_true=df.KS_pump_mm, y_pred=df.KS_netGW_mm, squared=False)
print('RMSE: ', rmse)

fig, ax = plt.subplots()
ax.scatter(df.KS_pump_mm, df.KS_netGW_mm, alpha=0.3, marker='.')
ax.set_xlabel('Actual pumpig (mm)')
ax.set_ylabel('Predicted pumpig (mm)')
ax.plot([0, 1], [0, 1], '-r', transform=ax.transAxes)
```
