```python
import sys

sys.path.append('H:/WestUS_WaterUSe/python_scripts')

import os
import numpy as np
import pandas as pd
from glob import glob
import rasterio as rio
import geopandas as gpd
from rasterio.plot import show
import matplotlib.pyplot as plt
from system_process import makedirs, copy_file
from raster_process import read_raster_arr_object, write_array_to_raster, clip_resample_reproject_raster
```

```python
WestUS_county = '../../Data/shapefiles/Western_US/WestUS_county.shp'
county_df = gpd.read_file(WestUS_county)
county_df['fips'] = county_df['STATEFP'] + county_df['COUNTYFP']
county_df['fips'] = county_df['fips'].astype('int64')
print(county_df.info())
county_df.plot(facecolor='green')
```

```python
wateruse_df = pd.read_csv('../../Data/USGS_water_use_data/WestUS_county_gw_use.csv')
wateruse_df = wateruse_df[wateruse_df['Year']==2015]
wateruse_df = wateruse_df.sort_values(by='fips', ascending=True)
wateruse_df.head()
```

```python
county_df = county_df.sort_values(by='fips', ascending=True)
```

```python
county_df = county_df.merge(wateruse_df, how='inner', on=['fips'])
county_df = county_df.sort_values(by='fips', ascending=True)  # sorting again to check values
county_df.head()
```

```python
county_df.shape
```

```python
state_df = gpd.read_file('../../Data/shapefiles/Western_US/WestUS_states.shp')

base = county_df[county_df['total_gw_observed']>30].plot(column='total_gw_observed', vmin=0, vmax=300)
state_df.plot(ax=base, facecolor='none', edgecolor='red')
```

```python
import dask.dataframe as dd
```

```python
train_set = dd.read_csv(r'H:\US_County_WaterUSe\Data\Compiled_data\predictor_csv\train_data.csv')
val_set = dd.read_csv(r'H:\US_County_WaterUSe\Data\Compiled_data\predictor_csv\validation_data.csv') 
test_set = dd.read_csv(r'H:\US_County_WaterUSe\Data\Compiled_data\predictor_csv\test_data.csv')

train_fips = train_set['fips'].compute().unique().tolist()
val_fips = val_set['fips'].compute().unique().tolist()
test_fips = test_set['fips'].compute().unique().tolist()
```

```python
# plotting counties using in train, validation, and test data

county_df_train = county_df[county_df['fips'].isin(train_fips)]
county_df_val = county_df[county_df['fips'].isin(val_fips)]
county_df_test = county_df[county_df['fips'].isin(test_fips)]


state_df = gpd.read_file('../../Data/shapefiles/Western_US/WestUS_states.shp')
fig, axs = plt.subplots(1, 3, figsize=(24, 24))

county_df_train.plot(column='gw_withdraw', vmin=0, vmax=300, ax=axs[0])
county_df_val.plot(column='gw_withdraw', vmin=0, vmax=300, ax=axs[1])
county_df_test.plot(column='gw_withdraw', vmin=0, vmax=300, ax=axs[2])

for i in range(3):
    state_df.plot(ax=axs[i], facecolor='none', edgecolor='red')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

train = county_df_train.plot(facecolor='lightyellow', ax=ax, legend=True)
val = county_df_val.plot(facecolor='lightblue', ax=ax, legend=True)
test = county_df_test.plot(facecolor='silver', ax=ax, legend=True)

state_df.plot(facecolor='none', edgecolor='black', ax=ax)
# ax.legend(loc='lower left', fontsize=12, labels=['train', 'val', 'test'])  # add legend correctly
```
