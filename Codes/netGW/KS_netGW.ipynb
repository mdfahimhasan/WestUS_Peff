```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv('../../scratch/netGW_KS/KS_netGW.csv')
df
```

```python
r2 = r2_score(df.KS_pump, df.KS_netGW)
print('R2: ', r2)

rmse = mean_squared_error(y_true=df.KS_pump, y_pred=df.KS_netGW, squared=False)
print('RMSE: ', rmse)

fig, ax = plt.subplots()
ax.scatter(df.KS_pump, df.KS_netGW, alpha=0.3, marker='.')
ax.plot([0, 1], [0, 1], '-r', transform=ax.transAxes)
```
