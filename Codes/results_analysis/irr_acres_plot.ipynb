```python
plot_save_path = './plots/irr_acres_diff_all.png'

basin_names = ('GMD4, KS (LANID + AIM-HPA)', 'GMD3, KS (LANID + AIM-HPA)', 'Republican River Basin, CO (IrrMapper)', 'Diamond Valley, NV (IrrMapper)')

make_scatter_plot_irr_area(df=df, x='Irr data m2', y='In situ m2', hue='basin_name',
                           fontsize=12, figsize=(8, 6), basin_labels=basin_names,
                           xlabel='Irrigated area from irrigation datasets (m$^2$)', 
                           ylabel='Irrigated area from in-situ records (m$^2$)', 
                           scientific_ticks=False, lim=(0, 9000000000),
                           savepath=plot_save_path, legend='on')
```
