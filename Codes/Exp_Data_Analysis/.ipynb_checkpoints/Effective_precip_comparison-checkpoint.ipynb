```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
from glob import glob
import contextily as cx
import geopandas as gpd
import matplotlib.pyplot as plt
```

#### Training zone 4

```python
point_shape = '../../Data_main/shapefiles/Effective_precip_comparison/Random_points_shape_updated/sampled_points_trainZone4.shp'
training_zone = '../../Data_main/shapefiles/Effective_precip_comparison/Training_zones_ROI/training_zone_4.shp'
westus_shape = '../../Data_main/shapefiles/Western_US_ref_shapes/WestUS_states.shp'

point_gdf_4 = gpd.read_file(point_shape)
zone_gdf_4 = gpd.read_file(training_zone)
westus_gdf = gpd.read_file(westus_shape)

# point_gdf_4.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4), gridspec_kw={'width_ratios': [3, 4]})

westus_gdf.plot(ax=ax[0], facecolor='none', edgecolor='black', alpha=0.5)
zone_gdf_4.plot(ax=ax[0], facecolor='none', edgecolor='blue')
cx.add_basemap(ax[0], source=cx.providers.OpenTopoMap, crs='EPSG:4269', alpha=0.8)
ax[0].set_axis_off()

sns.set_style('darkgrid')

x_axis = list(range(len(point_gdf_4)))
ax[1].scatter(x_axis, point_gdf_4['et_ensembl'], label='OpenET 30m', facecolor='green', edgecolor='white')
ax[1].scatter(x_axis, point_gdf_4['Ef_Pr_DK'], label='Eff Pr. DK', facecolor='blue',edgecolor='white')
ax[1].scatter(x_axis, point_gdf_4['Rain_ET'], label='Rainfed cropET 2km', facecolor='salmon',edgecolor='white')
ax[1].legend()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
sns.set_style('darkgrid')

x_axis = list(range(len(point_gdf_4)))
sns.boxplot(y=point_gdf_4['et_ensembl'], ax=ax[0], color='royalblue')
ax[0].set_ylabel('OpenET 30m (mm/month)')

sns.boxplot(y=point_gdf_4['Ef_Pr_DK'], ax=ax[1], color='tomato')
ax[1].set_ylabel('Ef. Precip D.K. (mm/month)')

sns.boxplot(y=point_gdf_4['Rain_ET'], ax=ax[2], color='lightgreen')
ax[2].set_ylabel('Rainfed cropET 2km (mm/month)')

plt.setp(ax, ylim=(0, 150))
plt.tight_layout()
```

--------------------------

In training zones 9, 10, 12, 14, the effective precip estimate from David Ketchum is systematically lower than OpenET 30m and Rainfec cropET 2km (filtered and area averaged from OpenET). Technically, all 3 datasets are based on openET and should have similar range of values. While oepnET 30m and rainfed cropET 2km have similar ranges of values, effective precip estimates seem to be of lower value ranges for all the training zones, especially for training zones 9, 10, 12, 14 (out of 7).
