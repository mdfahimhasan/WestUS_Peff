## making irrigated area plots for basins in KS (GMD3 and GMD4) and CO (Republican Basin)

```python
df = pd.read_csv('../../Data_main/results_eval/irr_acres/all_irr_acres.csv')
df.head()
```

```python
plot_save_path = './plots/irr_acres_diff_all.png'

basin_names = ('GMD3, KS', 'GMD4, KS', 'Republican Basin, CO', 'Diamond Valley, NV')

make_scatter_plot(df=df, x1='Irr data m2', y1='In situ m2', hue='basin_name',
                        fontsize=12, figsize=(5, 5), basin_labels=basin_names,
                        xlabel='Irrigated area from Irrigation dataset (m$^2$)', 
                        ylabel='Irrigated area from In situ records (m$^2$)', 
                        scientific_ticks=False, lim=(0, 9000000000),
                        savepath=plot_save_path)
```
