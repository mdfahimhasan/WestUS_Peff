```python
import calendar
import pandas as pd
import geopandas as gpd
```

# New GW Irrigation data for HUC12 from 2000-2020
__USGS source:__ https://www.sciencebase.gov/catalog/item/6488734cd34ef77fcafe347a

```python
# USGS GW Irrigation unfiltered dataset for CONUS
conus_df = pd.read_csv('../../Data_main/USGS_water_use_data/USGS_new_wateruse_data_HUC12/IR_HUC12_GW_2000_2020.csv')
conus_df
```

```python
# WestUS HUC12 shapefile
huc12_westUS = gpd.read_file('../../Data_main/shapefiles/HUCs/HUC12/HUC12WestUS.shp')
huc12_westUS.columns
```

```python
# HUC12 no from west US HUC12 shapefile
westUS_HUC12s = list(huc12_westUS['huc12'])
westUS_HUC12s[:5]
```

```python
# Unique HUC12 no. in USGS database
usgs_HUC12s = [i for i in conus_df.columns if i not in ['Year', 'Month']]
usgs_HUC12s[:5]
```

```python
# USGS HUC12 no. filtered with westUS HUC12 shapefile
usgs_HUC12s= set(usgs_HUC12s)
westUS_HUC12s = set(westUS_HUC12s)

usgs_HUC12s_filtered = list(usgs_HUC12s.intersection(westUS_HUC12s))
usgs_HUC12s_filtered[:5]
```

```python
# USGS HUC12 filtered for WestUS
usgs_HUC12s_filtered = ['Year', 'Month', ] + usgs_HUC12s_filtered
westUS_HUC12_df = conus_df[usgs_HUC12s_filtered]
westUS_HUC12_df[westUS_HUC12_df.isin([999, 888])] = 0   # setting null values to zero; no GW withdrawal

westUS_HUC12_df.head()
```

## Counting number of days in a month

```python
df = westUS_HUC12_df.copy()
```

```python
# # Counting number of days in months
df['num_days_month'] = df.apply(lambda row: calendar.monthrange(int(row['Year']), int(row['Month']))[1], axis=1)
# df.head()
```

##  Converting monthly mean GW withdrawal to monthly total GW withdrawal

```python
# converting monthly mean GW withdrawal to monthly total GW withdrawal
columns_temporal = ['Year', 'Month', 'num_days_month']
cols_to_multiply = [i for i in df.columns if i not in columns_temporal]

multiplied_cols = df[cols_to_multiply].mul(df['num_days_month'], axis=0)
df.update(multiplied_cols)
# df.head()
```

## Converting monthly totals to annual GW withdrawal

```python
# # summing for annual GW estimates
df = df.drop(columns=['Month', 'num_days_month'])
df_annual = df.groupby(by=['Year']).sum().reset_index()
df_annual
```

## Tranposing the columns as rows
The annual dataframe has HUC12 names as column names. But the WestUS HUC12 shapefile has HUC12 names in a column (as rows). We need to transpose the annual dataframe.

```python
df_annual_T = df_annual.T
df_annual_T.columns = df_annual_T.iloc[0].astype(int).astype(str)  # setting 1st row as column name
df_annual_T = df_annual_T.iloc[1:]  # dropping 1st row from dataframe
df_annual_T = df_annual_T.reset_index()
df_annual_T.columns.values[0] = 'huc12'
df_annual_T.head()
```

## Connecting the Annual GW USe dataframe with the HUC12 shapefile

```python
huc12_westUS_with_GW_irr = huc12_westUS.merge(df_annual_T, on='huc12')
huc12_westUS_with_GW_irr = huc12_westUS_with_GW_irr.drop(columns=['noncontr00', 'sourcedata',  'areaacres',  'humod', 'sourcefeat',
                                                                  'sourceorig', 'shape_area', 'shape_leng', 'hutype', 'tnmid', 'metasource',   
                                                                  'loaddate',    'gnis_id', 'noncontrib'])
huc12_westUS_with_GW_irr.head()
```

```python
huc12_westUS_with_GW_irr.columns
```

### removing HUC12s with no groundwater water use over 2000-2020

```python
gw_data_columns = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
                   '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
                   '2019', '2020']
```

```python
huc12_westUS_with_GW_irr = huc12_westUS_with_GW_irr[(huc12_westUS_with_GW_irr[gw_data_columns] != 0).any(axis=1)]
huc12_westUS_with_GW_irr.head()
```

## area calculation

```python
huc12_westUS_with_GW_AF = huc12_westUS_with_GW_irr
```

```python
# converting to projected coords
huc12_westUS_with_GW_AF = huc12_westUS_with_GW_AF.to_crs('EPSG:3857')
huc12_westUS_with_GW_AF = huc12_westUS_with_GW_AF.drop(columns=['areasqkm'])  # dropping because of some weird values
huc12_westUS_with_GW_AF['areasqkm'] = huc12_westUS_with_GW_AF['geometry'].area / (1000 * 1000) # unit in sqkm
```

## unit conversion from MGD to AF/year

```python
huc12_westUS_with_GW_AF[gw_data_columns] = huc12_westUS_with_GW_AF[gw_data_columns] * 3.0689 # MGD to AF/year
huc12_westUS_with_GW_AF.head()
```

## saving the HUC12 shapefiles with annual GW withdrawal 
`GW irrigation unit in AF/year`

```python
huc12_westUS_with_GW_AF.to_file('../../Data_main/USGS_water_use_data/USGS_new_wateruse_data_HUC12/HUC12_WestUS_with_Annual_GW.shp')
```
